```python
#¿como sacar de los datos del acelerometro los ciclos de la señal EMG?
#hay q coger los picos obtenidos en ta (acelerometro) y hacer una equivalencia en tm (emg marcha), pq el vector tiempo del EMG y del acelerometro son distintos.  tm del 0 al 139,9513235 y 5687 elementos y ta o al 139,98825 y 20741 elementos
# todo esto es para poder dividir las señales emg (TR TL, GL y GR) en ciclos, poder hacer la media y poder analizar los ciclos
# Respuesta: hay q interpolar EMG en los tiempos del acelerómetro
TR_interp = np.interp(ta, tm, TR)
TL_interp = np.interp(taL, tm, TL)
GR_interp = np.interp(ta, tm, GR)
GL_interp = np.interp(taL, tm, GL)
#ahora todo esta en el eje temporal ta
print(len(TR_interp))
print(len(TL_interp))
```

```python
N = 100  # % del ciclo al que normalizamos (de o a 100) = puntos normalizados por ciclo

def extraer_ciclos(signal, peaks, valid_idx):
    ciclos = []
    for i in valid_idx:
        ini = peaks[i]
        fin = peaks[i+1] #aqui se cojen los instantes de ta en que empieza y acaba cada uno de los ciclos válidos 
        ciclo = signal[ini:fin] 
        x_old = np.linspace(0, 1, len(ciclo))
        x_new = np.linspace(0, 1, N) #normaliza al 100
        ciclos.append(np.interp(x_new, x_old, ciclo)) 
    return np.array(ciclos)

TRc = extraer_ciclos(TR_interp, picosa, valid_idx)
TLc = extraer_ciclos(TL_interp, picosaL, valid_idxL)
GRc = extraer_ciclos(GR_interp, picosa, valid_idx)
GLc = extraer_ciclos(GL_interp, picosaL, valid_idxL)
#se obtiene un array con la señal EMG de todos los ciclos validos de cada uno de los musculos. len(TRc) es 75 pq dentro estan los 75 ciclos validos, cada uno de 100 puntos (pq ya estan normalizados)
print(len(TRc)) #esto es 75, que son los ciclos validos
print(len(TRc[0])) #esto es 100, que son los puntos normalizados de cada ciclo valido
print(len(TLc)) #esto es 81, que son los ciclos validos 
print(len(TLc[0])) #esto es 100, que son los puntos normalizados de cada ciclo valido
```

```python
TR_mean = np.mean(TRc, axis=0)
TR_std  = np.std(TRc, axis=0)

TL_mean = np.mean(TLc, axis=0)
TL_std  = np.std(TLc, axis=0)

GR_mean = np.mean(GRc, axis=0)
GR_std  = np.std(GRc, axis=0)

GL_mean = np.mean(GLc, axis=0)
GL_std  = np.std(GLc, axis=0)

x = np.linspace(0, 100, N)
print(len(x))
```

```python
#sin norm,  para la parte de metodologia del informe
plt.plot(x, TR_mean, label= 'Tibial R')
plt.fill_between(x, TR_mean-TR_std, TR_mean+TR_std, alpha=0.15)
plt.title('Ciclo de la señal EMG - velocidad ' + str(choice))
plt.xlabel('Porcentaje del ciclo (%)')
plt.ylabel('EMG (V)')
plt.xlim(1,100)
plt.ylim(0,0.0009)
plt.legend()
plt.show()
 
plt.plot(x, GR_mean, color ='green',label= 'Gemelo R')
plt.fill_between(x, GR_mean-GR_std, GR_mean+ GR_std,  color ='green', alpha=0.15)
plt.title('Ciclo de la señal EMG - velocidad ' + str(choice))
plt.xlabel('Porcentaje del ciclo (%)')
plt.ylabel(' EMG (V)')
plt.ylim(0,0.0009)
plt.xlim(1,100)
plt.legend()
plt.show()
```
